##TIME COMPLEXITY

Rate of increace in time with respect to input size.


Or in simple language:
How much time your code takes to run as input size increases.



**Example: Finding a name in a list**

You have a list of 5 names:

[Ram, Shyam, Riya, Aman, Neha]

You want to find Neha

You check one by one:

1st->Ram

2nd->Shyam

3rd->Riya

4th->Aman

5th->Neha (correct)->stop

-Took 5 steps

Now imagine list has 1000 names


Worst case:
You may check 1000 times


So:

Input = 5 -> 5 steps

Input = 1000 -> 1000 steps


Time increases with input size -> this is time complexity

**What is Big-O Notation?**

Big-O tells us the

* growth of time complexity

* It ignores small details and focuses on main behavior

###Rules
**1. Always calculate time complexity in terms of worst case.**
* worst case denoted as Big-Oh(O)

* Average case denoted as Theta

* Best case denoted as Omega

**2. Avoid the constant values.**

**3. Avoid lower bound**



```python
#1 worst case
age = 10

if age >= 80:                #1 operation
    print("Super Senior")

elif age >= 60 and age < 80: #2 operation
    print("Senior")

elif age >= 24 and age < 60: #3 operation
    print("Adult")

elif age >= 16 and age < 24: #4 operation
    print("Teen")

else:                        #5 operation
    print("Baby")
```

Count operations
Each if / elif is a comparison
In worst case -> all conditions are checked

Number of checks:

1st condition

2nd condition

3rd condition

4th condition
else

 Total ≈ constant number of steps (like 4-5 checks)



Avg case= (number of best case + number of worst case)/2

###Why do we always consider the worst case?
1. It gives a guarantee of performance

Worst case tells:

“No matter what input comes, the algorithm will never be slower than this”

2. Avoid Constant Values
Meaning

Ignore fixed numbers because they do not affect growth


```python
#Examples
n=5
for i in range(n):
    print(i)

for j in range(n):
    print(j)

#Steps = n + n = 2n
#O(2n) → O(n)
```

```python
for i in range(n):
    print(i)

print("Hello")

#Steps = n + 1
#O(n + 1) → O(n)
```

2. Avoid Lower Order Terms
Meaning

Keep only the highest growing term

```python
for i in range(n):
    print(i)

for j in range(n*n):
    print(j)

#Steps = n + n²
#O(n + n²) → O(n²)
```

O(1) < O(log n) < O(n) < O(n²) < O(n³)

* Always pick the highest

*Common Big-O Types*
1. O(1) — Constant Time (Best)

Always same time, no matter input

Example:

```python
arr = [1,2,3]
print(arr[0]) # one step
```

2. O(n) — Linear Time

```python
for i in arr:
    print(i) #if array size=n -> runs n times
```

3. O(n²) — Quadratic Time




```python
for i in arr:
    for j in arr:
        print(i, j) # If n = 3 -> 9 times
                    # If n = 100 -> 10,000 times
```

4. O(log n) — Fast (Binary Search)


Instead of checking one by one, we divide

Example:

Search in sorted list
Remove half every step

##Calculate Time Complexity

```python
for i in range(n+1):     #i=0, j= 0,2,3,4,...n
    for j in range(n+1): #i=1, j= 0,2,3,4,...n
        print(i, j)      #i=2, j= 0,2,3,4,...n
                         #i=n, j= 0,2,3,4,...n

#Outer loop = n
#Inner loop = n
#Total = n × n = n²

#TC=O(n²)
```

```python
for i in range(n+1):     #i=0, j= 0
    for j in range(i+1): #i=1, j= 0,1
        print(i, j)      #i=2, j= 0,1,2,
                         #i=n, j= 0,1,2,3,4,...n

#Operations=1 + 2 + 3 + 4 + ... + (n+1)
# n(n+1)/2

#TC=O(n²)
```

#SPACE COMPLEXITY

How much extra memory (space) your program uses as input size increases

* Total memory used by an algorithm as input size increases

* Represented using Big-O notation


**Types of Space**

**1. Auxiliary Space**

Extra space used by algorithm

Temporary variables
Data structures created inside code
Example
x = 5

y = 10

z = 15

total = x + y + z

Fixed variables -> no growth

**Auxiliary Space = O(1)**


**2. Input Space**

Memory used to store input data

**Example**
arr = [1, 2, 3, 4, 5]


If Array size = n

Input Space = O(n)


Total Space = Input Space + Auxiliary Space
