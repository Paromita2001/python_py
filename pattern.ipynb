```python
#square of side n
n=5
for i in range(n):
  for j in range(n):
    print('*',end='')
  print()
```

```python
#Hollow Square of side 'N' (N=5)
n=5
for i in range(n):
  for j in range(n):
    if j==0 or j==4 or i==0 or i==4:
      print('*', end= ' ')
    else:
      print('  ', end=' ')
  print()
```

```python
#Rectangle Pattern (rows=3, cols=6)
r=3
c=6
for i in range(r):
  for j in range(c):
    print('*', end=' ')
  print()
```

```python
#Right Angled Triangle (N=5)
n=5
for i in range(n):
  for j in range(i+1):
    print('*',end=' ')
  print()
```

```python
#Inverted Right Angled Triangle
n=5
for i in range(n):
  for j in range(n-i):
    print('*',end=' ')
  print()
```

```python
#Pyramid Pattern (N=5)
n = 5
for i in range(n):
    for j in range(n - i - 1):
        print("  ", end="")
    for k in range(2 * i + 1):
        print("*", end="")
    print()
```

```python
#Inverted Pyramid Pattern (N=5)
n = 5
for i in range(n, 0, -1):
    print("  " * (n - i) + "*" * (2 * i - 1))
```

```python
#Right Angled Triangle with Numbers (N=5)
n = 5
for i in range(n+1):
    for j in range(1,i+1):
        print(j, end=" ")
    print()
```

```python
#Floyd’s Triangle (N=5)
n = 5
num=1
for i in range(n+1):
    for j in range(1,i+1):
        print(num, end=" ")
        num+=1
    print()
```

```python
#Diamond Pattern (N=5)

n = 5
for i in range(n):
    print("  " * (n - i - 1) + "*" * (2 * i + 1))

for i in range(n - 2, -1, -1):
    print("  " * (n - i - 1) + "*" * (2 * i + 1))

```

```python
#Right Angled Triangle II (aligned right) (N=5)
n = 5
for i in range(1,n+1):
  print('  '*(n-i)+'*'*i)
```

```python
#Sandglass Pattern (N=5)
n = 5

for i in range(n, 0, -1):
    print("  " * (n - i) + "*" * (2 * i - 1))

for i in range(2, n + 1):
    print("  " * (n - i) + "*" * (2 * i - 1))
```

```python
#Hollow Right Triangle (N=5)
n = 5
for i in range(1, n + 1):
    for j in range(1, i + 1):
        if j == 1 or j == i or i == n:
            print("*", end=" ")
        else:
            print("  ", end=" ")
    print()
```

```python
n = 5
for i in range(n, 0, -1):
    for j in range(1, i + 1):
        if j == 1 or j == i or i == n:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
#Number Pyramid Pattern (N=5)
n = 5
for i in range(1, n + 1):
    print(" " * (n - i), end="")
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```
